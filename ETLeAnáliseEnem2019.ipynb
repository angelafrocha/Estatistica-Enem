```python
#Removendo colunas que não serão analisadas
df1 = df.drop(columns=['CO_MUNICIPIO_RESIDENCIA',
                       'CO_UF_RESIDENCIA',
                       'CO_MUNICIPIO_NASCIMENTO',
                       'CO_UF_NASCIMENTO',
                       'SG_UF_RESIDENCIA',
                       'NO_MUNICIPIO_NASCIMENTO',
                       'CO_UF_NASCIMENTO',
                       'SG_UF_NASCIMENTO',
                       'TP_ANO_CONCLUIU',
                       'TP_ENSINO',
                       'CO_MUNICIPIO_ESC',
                       'CO_UF_ESC',
                       'SG_UF_ESC',
                       'TP_DEPENDENCIA_ADM_ESC',
                       'TP_LOCALIZACAO_ESC',
                       'TP_SIT_FUNC_ESC',
                       'CO_ESCOLA'], )
```

```python
df1.head()
#foi observado aqui que as notas de CH,CN, Mat e Ling estão multilicadas
```

```python
df1.shape
```

```python
#Corrigindo as notas
df1.loc[:,'NU_NOTA_CN'] /= 10
df1.loc[:,'NU_NOTA_CH'] /= 10
df1.loc[:,'NU_NOTA_LC'] /= 10
df1.loc[:,'NU_NOTA_MT'] /= 10
```

```python
df1.head(3)
```

```python
#Renomeando colunas
df1 = df1.rename(columns={'NU_NOTA_REDACAO': 'NOTA_REDACAO',
                          'NU_NOTA_CN': 'NOTA_CN',
                          'NU_NOTA_CH': 'NOTA_CH',
                          'NU_NOTA_LC': 'NOTA_LC',
                          'NU_NOTA_MT': 'NOTA_MT',
                          'NU_NOTA_COMP1': 'COMP1',
                          'NU_NOTA_COMP2': 'COMP2',
                          'NU_NOTA_COMP3': 'COMP3',
                          'NU_NOTA_COMP4': 'COMP4',
                          'NU_NOTA_COMP5':'COMP5',
                          'NU_IDADE': 'IDADE',
                          'TP_SEXO': 'SEXO',
                          'TP_COR_RACA': 'RACA',
                          'Q025': 'INTERNET',
                          'TP_ESCOLA':'ESCOLA'})
```

```python
df1.head(3)
```

####Renomeando dados das colunas para melhor entendimento

```python
df1['RACA'].unique()
```

```python
df1["RACA"] = df1["RACA"].replace({0:"nao_declarado", 
                                   1:"branca", 
                                   2:"preta", 
                                   3:"parda", 
                                   4:"amarela", 
                                   5:"indigena"})
```

```python
df1["TP_LINGUA"] = df1["TP_LINGUA"].replace({0:"ingles", 
                                             1:"espanhol"})
```

```python
df1["ESCOLA"] = df1["ESCOLA"].replace({1:"nao_respondeu", 
                                       2:"publica", 
                                       3:"privada", 
                                       4:"exterior"})
```

```python
df1["INTERNET"] = df1["INTERNET"].replace({"A":"nao", 
                                           "B":"sim"})
```

```python
df1['TP_ESTADO_CIVIL'] = df1['TP_ESTADO_CIVIL'].replace({0:'nao_informado',
                                                         1:'solterio',
                                                         2:'casado/uniao_est',
                                                         3:'divorciado/separado',
                                                         4:'viuvo'})
```

```python
df1['TP_NACIONALIDADE'] = df1['TP_NACIONALIDADE'].replace({0:'nao_informado',
                                                           1:'brasileiro',
                                                           2:'br_naturalizado',
                                                           3:'estrangeiro',
                                                           4:'br_nato_nasc_esterior'})
```

```python
df1.head()
```

###Análise Exploratória e Transformação dos dados

Análise das Idades

```python
df1['IDADE'].value_counts()
```

```python
#verificando os valores únicos
np.sort(pd.unique(df1['IDADE']))
#idades abaixo de 12 anos são suspeitas(alta probabilidade de erro)
```

```python
df1.shape
```

```python
#criando novo df apenas com >= 12 anos
df_maior11 = df1.loc[df1.IDADE > 11]
```

```python
#Foram desconsideradas apenas 16 linhas, menos de 2%
df_maior11.shape
```

```python
np.sort(pd.unique(df_maior11['IDADE']))
```

#####Treineiros

```python
#verificando a quantidade de treineiros
df_maior11.query('IN_TREINEIRO == 1')['IN_TREINEIRO'].value_counts()
```

```python
#criando um df apenas com os dados dos treineiros
df_treineiros = df_maior11.loc[df_maior11.IN_TREINEIRO == 1]
```

```python
df_treineiros.head(3)
```

```python
#salvando df_treineiros para análise posterior
df_treineiros.to_csv('treineiros_enem_2019_sp.csv', encoding = 'iso-8859-1', index = False)
```

```python
# Separação entre os treineiros e os vestibulandos
df_vestibulandos = df_maior11.loc[df_maior11.IN_TREINEIRO == 0]
```
